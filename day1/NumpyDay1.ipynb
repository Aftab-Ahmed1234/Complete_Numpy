```python
# N_D_Array
import numpy as np
arr=np.array([1,2,3,4,5])
print(arr)
print(type(arr))
```

```python
# Use a tuple to create a NumPy array:
import numpy as np
arr=np.array((1,2,3,4,5))
print(arr)
```

```python
# 0D Araay
import numpy as np
arr=np.array(11)
print(arr)

print("Dimensions: ",arr.ndim)
```

```python
# One-D Array 
import numpy as np 
arr=np.array([1,2,3,4,5])
print(arr)
print("Dimensions: ", arr.ndim)
```

```python
# 2-D Array
import numpy as np
arr=np.array([[1,2,3,4,5],[2,4,6,8,10]])
print(arr)
print("Dimensions: ",arr.ndim)
```

```python
# 3-D Array 
import numpy as np 
arr=np.array([[[1,2,3,4,5],[2,4,6,8,10],[1,3,5,7,9]]])
print(arr)
print(arr.ndim)
```
