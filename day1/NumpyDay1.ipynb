```python
# One-D Array 
import numpy as np 
arr=np.array([1,2,3,4,5])
print(arr)
print("Dimensions: ", arr.ndim)
```

```python
# 2-D Array
import numpy as np
arr=np.array([[1,2,3,4,5],[2,4,6,8,10]])
print(arr)
print("Dimensions: ",arr.ndim)
```

```python
# 3-D Array 
import numpy as np 
arr=np.array([[[1,2,3,4,5],[2,4,6,8,10],[1,3,5,7,9]]])
print(arr)
print(arr.ndim)
```

```python
# Array Indexing
# get first element
import numpy as np 
arr=np.array([1,2,4,5])
print(arr[0])
```

```python
# get second element
import numpy as np 
arr=np.array([1,2,3,4,5])
print(arr[1])
```

```python
#get element on 3rd and 4rth positon
import numpy as np 
arr= np.array([1,2,3,4,5])
print("Element on third position",arr[2])
print("Element on fourt position",arr[3])
```

```python
#get second element from 1st row
import numpy as np
arr=np.array([[2,4,6,8],[1,3,5,7]])
print(arr)
print("2nd element on 1st row is:  ",arr[0][1])
```

```python
#5th element on the second row
import numpy as np 
arr=np.array([[2,4,6,8,10],[1,3,5,7,9]])
print(arr)
print("5th element on the 2nd position is: ", arr[1][4])
print("Second way of printing: ", arr[1,4])
```

```python
# 3-D array 
# access 3rd element of the second array of the first array 
import numpy as np 
arr=np.array([[[1,2,3],[4,5,6]],[[2,4,6],[1,3,5]]])
print(arr[0,1,2])
```

```python
#Print Last Element Of the one 1-D array
import numpy as np
arr= np.array([[1,2,3],[2,4,6]])
print("Last Elenment of the 1-D",arr[1,-1])
```

```python
# start from index 0 to 4 and it takes one step by each element
import numpy as np 
arr=np.array([1,2,3,4,5])
print(arr[0:4:1])
```

```python
# start from index 0 to 4 and it takes two step by each element
import numpy as np 
arr=np.array([1,2,3,4,5])
print(arr[0:4:2])
```

```python
#it Start form 0 index and ends on 4 index
import numpy as np
arr=np.array([2,4,6,8])
print(arr[0:4])
```

```python
# Slice element form 3 index end of the array
import numpy as np 
arr=np.array([1,2,3,4,5])
print(arr[3:])
```

```python
#Slice element form 0 index to 3(not included)
import numpy as np 
arr=np.array([1,2,3,4,5])
print(arr[:3])
```

```python
#Slice element form -5 to end 
import numpy as np
arr=np.array([1,2,3,4,5])
print(arr[-5:])
```

```python
#Slice element form index -5 to end before index -1
import numpy as np 
arr=np.array([1,2,3,4,5,6,7])
print(arr[-5:-1])
```
